```python
# import libraries to plot data (2D)
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

path_lcs_users = '../../formatted_datasets/lcs/cresci-2017_genuine_accounts.csv'
path_lcs_bots1 = '../../formatted_datasets/lcs/cresci-2017_social_spambots_1.csv'
path_lcs_mixed = '../../formatted_datasets/lcs/mixed1.csv'

# load datasets in pandas dataframes
lcs_users = pd.read_csv(path_lcs_users)
lcs_bots1 = pd.read_csv(path_lcs_bots1)
lcs_mixed = pd.read_csv(path_lcs_mixed)

# There are 3 columns k,length,path. just remove path
lcs_users = lcs_users.drop(columns=['path'])
lcs_bots1 = lcs_bots1.drop(columns=['path'])
lcs_mixed = lcs_mixed.drop(columns=['path'])

```

```python
def calculate_derivative(dataframe):
    # Tri du DataFrame par la colonne 'k' pour s'assurer que les valeurs sont ordonnées
    dataframe = dataframe.sort_values(by='k')
    
    # Calcul de la dérivée
    derivative = dataframe['length'].diff() / dataframe['k'].diff()
    
    # Ajout de la colonne dérivée au DataFrame
    dataframe['derivative'] = derivative
    
    return dataframe
```

```python
#for genuine accounts and bots1, plot k on the x-axis and length on the y-axis. not with a scatter but with a line
plt.plot(lcs_users['k'], lcs_users['length'], label='Genuine Accounts')
plt.plot(lcs_bots1['k'], lcs_bots1['length'], label='Social Spambots 1')
plt.xlabel('k')
plt.ylabel('Length')
plt.title('LCS')
plt.legend()
plt.show()
```

```python
# dérivée de bots1 avec la fonction calculate_derivative
lcs_bots1_d = calculate_derivative(lcs_bots1)
```

```python
# plot de la dérivée de bots1
plt.plot(lcs_bots1_d['k'], lcs_bots1_d['derivative'], label='Social Spambots 1')
plt.xlabel('k')
plt.ylabel('Derivative')
plt.title('LCS Derivative')
plt.legend()
plt.show()
```

```python
# Plot only mixed with max y being 3000
plt.plot(lcs_mixed['k'], lcs_mixed['length'], label='Mixed')
plt.xlabel('k')
plt.ylabel('Length')
plt.title('LCS')
plt.legend()
plt.show()
```
